```python
df_new = df.drop(["casual", "registered"], axis = 1)
weather_columns = ['weather']
df_clean = pd.get_dummies(df_new,columns=weather_columns,drop_first=True)

df_clean.head()
```

```python
def feature_engineer(df):
    df["datetime"] = pd.to_datetime(df["datetime"])
    df["hour"] = df['datetime'].dt.hour
    #df["dayofweek"] = df['datetime'].dt.dayofweek
    df["month"] = df['datetime'].dt.month
    season_column = ['season']
    df_merged = pd.get_dummies(df,columns=season_column,drop_first=True)
    X_engineered  = df_merged.drop(["temp", "datetime"], axis = 1)

    return X_engineered
```

```python
y = df_clean["count"]

X = df_clean.drop("count", axis = 1)

X_train, X_test, y_train, y_test = train_test_split(X, y)

X_test.head()
```

```python
X_train_final = feature_engineer(X_train)
```

```python
k = RandomForestRegressor(max_depth = 12)
k.fit(X_train_final, y_train)
```

```python
X_test_final = feature_engineer(X_test)
```

```python
ypred = k.predict(X_test_final)
```

```python
np.sqrt(mean_squared_log_error(y_test, ypred))
```
