```python
#Which brands have the highest prices?

most_expensive_shoes = df.groupby('brand')[['prices.amountMax']].mean().sort_values(by = ['prices.amountMax'], ascending=False)
```

```python
most_expensive_shoes.plot()
```

```python
most_expensive_shoes = most_expensive_shoes[most_expensive_shoes["prices.amountMax"] > 150]
```

```python
most_expensive_shoes.plot.bar()
plt.title("Most expensive shoe brands")
```

```python
average_price_df = df.dropna(axis=0, subset=['prices.amountMax'])
```

```python
df.isnull().any()
```

```python
sliced_average_price_df = average_price_df[['brand', 'prices.amountMin','prices.amountMax']]
sliced_average_price_df['average_price'] = (sliced_average_price_df['prices.amountMin'] + sliced_average_price_df['prices.amountMax'])/2
sliced_average_price_df
```

```python
average_price_per_brand = sliced_average_price_df.groupby('brand')[['average_price']].mean().sort_values(by = ['average_price'])
average_price_per_brand
```

```python
import pylab as plt
average_price_per_brand.hist()
plt.title("Average prices of shoes per brand") 
plt.savefig("Average prices of shoes")
plt.figure(figsize=(50,50))
plt.axis('off')


```
