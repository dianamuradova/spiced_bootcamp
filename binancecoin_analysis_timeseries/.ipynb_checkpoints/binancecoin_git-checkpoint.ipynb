## The model is build to forecast the price of Binance Coin.  

### Initial data included market data and social media data obtained through CoinGecko API, later interpolated to hourly data

### Initial model is linear regression, then FB prophet, and finally ARIMA (please note: not included in the notebok are naive models)

```python
#getting data for binance coin social media stats 

import requests
import pandas as pd

timestamps = []
k_timestamps = []
from datetime import datetime

for i in range(682):
    t = 1505520000 + i*86400
    k = datetime.utcfromtimestamp(t).strftime('%d-%m-%Y')
    timestamps.append((t))
    k_timestamps.append(k)
     
dfs = [] 
    
for k in k_timestamps:
    m = k
    url = f'https://api.coingecko.com/api/v3/coins/binancecoin/history?date={k}&localization=false'
    result = requests.get(url)
    result_json = result.json()
    if 'community_data' not in result_json:
        print(k)
    else:
        community_data = pd.DataFrame(result_json['community_data'], index = [m])
        dfs.append(community_data)
        new_list_concat  = pd.concat(dfs)
    
new_list_concat["datetime"] = timestamps  
```

```python
new_list_concat_csv = new_list_concat.to_csv("binancecoin.csv")

#two dates are missing 09-10-2017 and 24-10-2017
```

```python
import pandas as pd
market_range = pd.read_csv("market_range.csv")
```

```python
market_binancecoin = market_range[market_range["coinname"] == "binancecoin"] #filtering for only binancecoin 
```

```python
market_binancecoin["date"] = pd.to_datetime(market_binancecoin["timestamp"], unit = "ms") #converting date column to datetime
```

```python
market_binancecoin['Year'] = market_binancecoin['date'].dt.year
market_binancecoin['Month'] = market_binancecoin['date'].dt.month
market_binancecoin['Day'] = market_binancecoin['date'].dt.day
market_binancecoin['Hour'] = market_binancecoin['date'].dt.hour
market_binancecoin['Weekday'] = market_binancecoin['date'].dt.weekday

market_binancecoin = market_binancecoin.drop(['timestamp'], axis = 1)

market_binancecoin = market_binancecoin.set_index('date')

market_binancecoin.head()
```

```python
social_binancecoin = pd.read_csv("binancecoin.csv", index_col = 0)
```

```python
merged_data = market_binancecoin.merge(social_binancecoin, left_index = True, right_index = True, how = 'left')
```

```python
merged_data_usd = merged_data[merged_data["currencyname"] == "usd"]
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline
plt.figure(figsize = (15, 10))
merged_data_usd["prices"].plot()
```

### Decided to include in analysis only twitter_users (as other social media use was highly correlated) and reddit_average_posts_48h

```python
features = ["prices", "market_caps", "total_volumes", "twitter_followers", "reddit_average_posts_48h"]
```

```python
intermediate_usd = merged_data_usd[features]
```

```python
intermediate_usd["twitter_followers_diff"] = intermediate_usd["twitter_followers"].diff()
```

```python
intermediate_usd["twitter_followers_diff"].plot()
```

```python
intermediate_usd.drop(["twitter_followers"], axis = 1, inplace = True)
```

```python
df_arima = intermediate_usd.copy()
```

```python
df_var_multi = intermediate_usd.copy()
```

```python
my_list = ["market_caps", "total_volumes", "reddit_average_posts_48h", "twitter_followers_diff"]

for l in my_list:
    for i in range(5):
        intermediate_usd[l + "" + f' {i}_shift'] = intermediate_usd[l].shift(i)
```

```python
intermediate_usd.isna().sum()
intermediate_usd.dropna(inplace = True)
```

```python
features_train = intermediate_usd.drop(['prices'], axis = 1).columns
```

```python
X_train = intermediate_usd[4320:14500][features_train] #decided to exclude the initial timestamps as the coin was just being established
X_test = intermediate_usd[14500:][features_train]
y_train = intermediate_usd[4320:14500]["prices"]
y_test = intermediate_usd[14500:]["prices"]
```

```python
y_train.plot()
```

```python
from sklearn.linear_model import Lasso
m = Lasso(alpha = 100.0)
m.fit(X_train, y_train)
m.coef_
```

```python
ypred = m.predict(X_train)
ypred_test = m.predict(X_test)
```

```python
plt.plot(X_train.index,ypred)
plt.plot(y_train)
```

```python
plt.plot(X_test.index, y_test)
plt.plot(X_test.index, ypred_test)
```

```python
m.score(X_train, y_train), m.score(X_test, y_test)
```

### FB prophet

```python
from fbprophet import Prophet
df_prophet = merged_data_usd[features]
df_prophet.reset_index(inplace = True)
df_prophet.rename(columns = {"index" : "ds", "prices" : "y"}, inplace = True)
```

```python
m = Prophet()
m.fit(df_prophet)
future = m.make_future_dataframe(periods = 240)
forecast = m.predict(future)
```

```python
fig1 = m.plot(forecast)
```

```python
fig2 = m.plot_components(forecast)
```

```python
y_train_arima = df_arima[:16000]["prices"]
y_test_arima = df_arima[16000:]["prices"]
y_train_arima.shape, y_test_arima.shape
```

```python
from statsmodels.tsa.arima_model import ARIMA
import numpy as np
m = ARIMA(y_train_arima, (3,1,1))
m.initialize()
r = m.fit()
```

```python
r.params
```

```python
ypred, err, conf = r.forecast(steps = 369) 
```

```python
plt.figure(figsize = (12, 10))
plt.plot(range(16000), y_train_arima, color = 'blue')
plt.plot(range(16000, 16369), y_test_arima, color = 'black')
plt.plot(range(16000, 16369), ypred, color = 'red', marker = 'o')
```
